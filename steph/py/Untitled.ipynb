```python
n = 225
k = 15
p = 0.25

w_min = 1
w_max = 20
```

```python
G = nx.erdos_renyi_graph(n, p)
while not nx.nx.is_connected(G):
    G = nx.erdos_renyi_graph(n, p)
```

```python
for (u, v, w) in G.edges(data=True):
    w['weight'] = random.randint(w_min, w_max)
```

```python
#nx.draw_networkx(G)
```

```python
sp = nx.shortest_path_length(G, weight='weight')
sp
```

```python
df = pd.DataFrame([(u, sorted(others.items())) for (u, others) in sp.items()], columns=['vertex', 'sortest paths'])
df
```

```python
S = sorted(random.sample(G.nodes(), k))
print(S)
```

```python
P = {}
for u in [u for u in G.nodes() if u not in S]:
    sp_S = {v: d for (v, d) in sp[u].items() if v in S}
    min_in_S = min(sp_S.keys(), key=(lambda k: sp_S[k]))
    P[u] = (min_in_S, sp[u][min_in_S])
df_P = pd.DataFrame([[u, data[0], data[1]] for (u, data) in P.items()], columns=['vertex', 'representative', 'sortest path weight'])
print(P)
print(df_P)
```

```python
Q = {}
for u in S:
    Q[u] = {v: d for (v, d) in sp[u].items() if v in S}
df_Q = pd.DataFrame([(u, sorted(others.items())) for (u, others) in Q.items()], columns=['vertex', 'sortest paths'])
print(df_Q)
```

```python
V = list(sorted(G.nodes()))
res = {}
for (u, v) in [(V[i], V[j]) for i in range(len(V)) for j in range(i+1,len(V))]:
    if u in S:
        (Su, du) = (u, 0)
    else:
        (Su, du) = P[u]
    if v in S:
        (Sv, dv) = (v, 0)
    else:
        (Sv, dv) = P[v]
    dS = Q[Su][Sv]
    res[(u, v)] = (sp[u][v], du + dS + dv)

df_res = pd.DataFrame([(u, v, data[0], data[1]) for ((u, v), data) in res.items()], 
                      columns=['from vertex', 'to_vertex', 'shortest path weight', 'upper bound'])

print(df_res)
```

```python
df_res['ratio'] = df_res['upper bound'] / df_res['shortest path weight']
#df_res['ratio'].plot(kind='bar')
```
